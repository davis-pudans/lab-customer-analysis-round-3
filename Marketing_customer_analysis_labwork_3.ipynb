```python
plt.figure(figsize = (10, 10))
plt.scatter(x=file1['total_claim_amount'], y=file1['income']) 
plt.xlabel("TTL Claim AMT")
plt.ylabel("Income")
plt.title("Income vs Total claim amount")
plt.show()
```

```python
plt.figure(figsize = (10, 10))
plt.plot('total_claim_amount','income',"",data=file1,linestyle='', marker='o', markersize=3,alpha=0.1, color="purple") 
plt.xlabel("TTL Claim AMT")
plt.ylabel("Income")
plt.title("Income vs Total claim amount")
plt.show()
```

```python
plt.figure(figsize = (10, 10))
sns.kdeplot(data = file1, x="total_claim_amount", y="income", cmap="Reds", shade=True)
plt.xlabel("TTL Claim AMT")
plt.ylabel("Income")
plt.title("Income vs Total claim amount")
plt.show()
```
