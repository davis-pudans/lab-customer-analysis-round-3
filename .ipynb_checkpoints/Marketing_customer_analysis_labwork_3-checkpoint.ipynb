```python
plt.figure(figsize = (10, 10))
sns.kdeplot(data = file1, x="total_claim_amount", y="income", cmap="Reds", shade=True)
plt.xlabel("TTL Claim AMT")
plt.ylabel("Income")
plt.title("Income vs Total claim amount")
plt.show()
```
